## Clean up the data
Did you notice anything unusual about the "house" amounts? 

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Slice the data  
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2) 
```

# New work here

## Drop unimportant columns

```python
df_clean = df_clean.drop(columns=['id'])
```

## Drop Nulls

```python
df_clean.dropna()
```

## Income Boxplot

```python
fig, ax = plt.subplots(figsize=(10,6))
df_clean.boxplot(column='income', ax=ax)
plt.show()
```

## Reported Satisfaction barchart

```python
df_clean.reported_satisfaction.value_counts(dropna=True).plot(kind='bar')
```

```python
df_clean.info()
```

Plot correlations

```python
sns.pairplot(df_clean)
```

Show correlations

```python
df_corr = df_clean.corr()
df_corr
```

Heatmap to see correlation better

```python
sns.heatmap(df_corr, 
            annot=True,
            cmap=sns.color_palette("vlag"), 
        xticklabels=df_corr.columns,
        yticklabels=df_corr.columns)

plt.plot()
```

It appears that handset price and income are highly positively correlated. There are some other correlations that are also high, but more intuitive such as overage and over 15mins calls per month.
