## Clean up the data
Did you notice anything unusual about the "house" amounts? 

How about the handset price and income?

```python
df_clean = df[(df['quantity'] >= 0)]
```

```python
df_clean['product_price'] = df['product_price'].str.replace(',','')
df_clean['product_price'] = df['product_price'].str.replace('$','')
df_clean.info()
```

```python
df_clean['total_amount'] = df['total_amount'].str.replace(',','')
df_clean['total_amount'] = df['total_amount'].str.replace('$','')
df_clean
```

```python
df_clean['total_amount'] = df_clean['total_amount'].apply(lambda x: float(x.split()[0].replace('$', '')))
df_clean['total_amount'] = df_clean['total']
```

```python

df_clean = df_clean.drop(columns=[''])
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```
