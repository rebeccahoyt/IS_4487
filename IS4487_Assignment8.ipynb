##Remove rows with NAs

```python

df_clean = df.dropna()
```

##Remove outlier

```python
# Remove the row with the single outlier in num_accts
df_clean = df_clean[df_clean.num_accts < df_clean.num_accts.quantile(0.99)]
```

##Remove negative values

```python
# Remove the rows with negative income values
df_clean = df_clean[df_clean.income >= 0]
```

##Convert variables into factors

```python
# Convert variables into factors
df_clean['job'] = pd.Categorical(df_clean['job'])
df_clean['product'] = pd.Categorical(df_clean['product'])
df_clean['chk_acct'] = pd.Categorical(df_clean['chk_acct'])
df_clean['sav_acct'] = pd.Categorical(df_clean['sav_acct'])
```
